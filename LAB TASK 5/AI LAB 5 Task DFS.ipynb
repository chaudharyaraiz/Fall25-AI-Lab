```python
def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=" ")
    for nxt in graph[start] - visited:
        dfs(graph, nxt, visited)
    return visited

graph = {
    '0': {'1', '2'},
    '1': {'0', '3', '4'},
    '2': {'0'},
    '3': {'1'},
    '4': {'2', '3'}
}

print("DFS Traversal (Graph):")
dfs(graph, '0')


class Node:
    def __init__(self, data):
        self.data = data
        self.left = None
        self.right = None

def inorder(root):       
    if root:
        inorder(root.left)
        print(root.data, end=" ")
        inorder(root.right)

def preorder(root):      
    if root:
        print(root.data, end=" ")
        preorder(root.left)
        preorder(root.right)

def postorder(root):     
    if root:
        postorder(root.left)
        postorder(root.right)
        print(root.data, end=" ")


root = Node(1)
root.left = Node(2)
root.right = Node(3)
root.left.left = Node(4)
root.left.right = Node(5)

print("\n\nTree Traversals:")
print("Inorder: ")
inorder(root)
print("\nPreorder: ")
preorder(root)
print("\nPostorder: ")
postorder(root)
```

```python

def dfs_stack(graph, start):
    visited, stack = set(), [start]
    while stack:
        node = stack.pop()
        if node not in visited:
            print(node, end=" ")
            visited.add(node)
            stack.extend(graph[node] - visited)

graph = {
    '0': {'1','2'},
    '1': {'0','3','4'},
    '2': {'0'},
    '3': {'1'},
    '4': {'2','3'}
}
print("DFS (Graph):")
dfs_stack(graph, '0')


class Node:
    def __init__(self,data):
        self.data=data
        self.left=None
        self.right=None

def dfs_tree(root):
    if not root: return
    stack=[root]
    while stack:
        node=stack.pop()
        print(node.data, end=" ")
        if node.right: stack.append(node.right)
        if node.left: stack.append(node.left)

root=Node(1)
root.left=Node(2)
root.right=Node(3)
root.left.left=Node(4)
root.left.right=Node(5)

print("\nDFS (Tree):")
dfs_tree(root)
```
