```python
import pandas as pd
import numpy as np
```

```python
dataset = pd.read_csv("Iris.csv")
dataset.head()
dataset.tail()
```

```python
print("We have {} rows.".format(dataset.shape[0]))
print("We have {} columns.".format(dataset.shape[1]))
```

```python
np.sum(pd.isnull(dataset))
```

```python
dataset = dataset.fillna(dataset.mean(numeric_only=True))
dataset = dataset.fillna(dataset.mode().iloc[0])
```

```python
dataset["Species"].unique()
```

```python
dataset["Id"] = dataset["Id"].astype(np.int64)
```

```python
dataset.drop("Id", axis=1, inplace=True)
```

```python
dataset.dtypes
```

```python
X = dataset.iloc[:, 0:-1]
y = dataset.iloc[:, -1]

print("X shape:", X.shape)
print("y shape:", y.shape)
```

```python
cat_columns = X.select_dtypes(['object']).columns
X[cat_columns] = X[cat_columns].apply(lambda col: pd.factorize(col)[0])
```
